```python
        indicator_categorical = indicator_data.copy()
        columns = indicator_categorical.columns[1:-1]
        def func(value):
            if 1.0 >= value >= 0.75:
                return 'H'
            elif 0.75 > value >= 0.50:
                return 'M'
            elif 0.50 > value >= 0.00:
                return 'L'
            else:
                raise ValueError("Data contains values out of range [1.0, 0.0]")
        try:
            indicator_categorical.loc[:, columns] = indicator_categorical.loc[:, columns].applymap(func)
        except ValueError as e:
            print("DEBUG:")
            print(f"  columns: {columns}")
            print("DATA with ERRORS == True:")
            print(indicator_categorical.loc[:, columns] < 0.0 | indicator_categorical.loc[:, columns] > 1.0)
```
